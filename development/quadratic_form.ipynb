```python
import numpy as np
import matplotlib.pyplot as plt

# Define the quadratic form parameters
A = np.array([[2, 1], [1, 2]])  # Symmetric matrix
b = np.array([1, -1])
c = 0

# Modified quadratic form function to handle arrays
def quadratic_form(x, y):
    # Compute the quadratic form for each (x, y) pair in the meshgrid
    return A[0, 0]*x**2 + (A[0, 1] + A[1, 0])*x*y + A[1, 1]*y**2 + b[0]*x + b[1]*y - c

# Generate a grid of x, y values
x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)

# Compute the quadratic form over the grid
Z = quadratic_form(X, Y)

# Plot the level curves
plt.contour(X, Y, Z, levels=[0], colors='blue')  # Level curve for f(x, y) = 0
plt.xlabel('x')
plt.ylabel('y')
plt.title('Level Curve of the Quadratic Form $x^TAx + b^Tx = c$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.show()
```
